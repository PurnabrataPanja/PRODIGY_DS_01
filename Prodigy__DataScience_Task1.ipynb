```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data from CSV file using pandas
filename = "population_distribution.csv"
data = pd.read_csv(filename)
data
```

```python
# Extract relevant columns
years = data['Year']
age_groups = data.columns[1:]
age_groups
```

```python
# Remove '%' from data and convert to float
for age_group in age_groups:
    data[age_group] = data[age_group].str.rstrip('%').astype(float)

data
```

```python
# Create a bar chart for each age group
plt.figure(figsize=(12, 8))
for i, age_group in enumerate(age_groups):
    plt.bar([x + i * 0.2 for x in years], data[age_group], width=0.2, label=age_group)

plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Percentage of Population by Age Group')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.xticks(years)
plt.tight_layout()
plt.show()
```

```python
# Create a bar chart for each age group
plt.figure(figsize=(12, 8))
for age_group in age_groups:
    plt.plot(years, data[age_group], label=age_group)

plt.xlabel('Year')
plt.ylabel('Percentage')
plt.title('Percentage of Population by Age Group')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
